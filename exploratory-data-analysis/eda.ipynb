```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/taiwanese-bankruptcy-prediction.csv')

df
```

```python
df.info()
```

```python
df_notnull = df.notnull()

df_notnull.to_csv('df-notnull.csv')

df_notnull
```

```python
df_sorted = df.reindex(sorted(df.columns), axis=1)

df_sorted.to_csv('df-sort-cols.csv')

df_sorted
```

```python
describe = df.describe()

describe.to_csv('df-describe.csv')

describe
```

```python
describe_sorted = describe.reindex(sorted(describe.columns), axis=1)

describe_sorted.to_csv('df-describe-sort-cols.csv')

describe_sorted
```
